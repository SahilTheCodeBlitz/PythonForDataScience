pandas data structure

pandas has two types of data structure 

1. series  2. dataframe 

series = its one dimensional array with indexes , it stores a single column of row of data in a dataframe. eg dff['Train No']

dataframe = its tabular spreadsheet like structure representing rows each of which contains one or multiple columns. eg entire dataset dff which contains rows and columns

also 
series = 1 dimesnional array capable of holding any type of data
dataframe = 2 dimesnional data structure with columns of potentially different data types
